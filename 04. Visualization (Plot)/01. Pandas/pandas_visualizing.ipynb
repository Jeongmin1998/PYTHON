## 1. Plot 그래프

```
df = pd.read_csv('https://bit.ly/ds-house-price-clean')
```

kind 옵션을 통해 원하는 그래프를 그릴 수 있다.

kind 옵션:
* line: 선그래프
* bar: 바 그래프
* barh: 수평 바 그래프
* hist: 히스토그램
* kde: 커널 밀도 그래프
* hexbin: 고밀도 산점도 그래프
* box: 박스 플롯
* area: 면적 그래프
* pie: 파이 그래프
* scatter: 산점도 그래프


데이터의 성격에 맞는 그래프 형식이 있다.

## line 그래프

* line 그래프는 데이터가 연속적인 경우 사용하기 적절 (예를 들면, 주가 데이터)

```
df['분양가'].plot(kind='line') #디폴트
#딱히 적절하지 않은 그래프
```

```
df_seoul = df.loc[df['지역'] == '서울']
```

```
df_seoul
```

```
df_seoul_year = df_seoul.groupby('연도').mean()
```

```
df_seoul_year
```

```
df_seoul_year.plot(kind='line')
```

## bar 그래프

bar 그래프는 그룹별로 비교할 때 유용

```
df.groupby('지역')['분양가'].mean()
```

```
df.groupby('지역')['분양가'].mean().plot(kind='bar')
```

```
df.groupby('지역')['분양가'].mean().plot(kind='barh')
```

## 히스토그램 (hist)

히스토그램은 **분포-빈도 를 시각화**

가로축에는 분포를, 세로축에는 빈도

```
df['분양가'].plot(kind='hist')
#2000-4000에 몰려있음
```

## 커널 밀도 그래프

* 히스토그램과 유사하게 밀도를 보여주는 그래프입니다.
* 히스토그램과 유사한 모양새를 갖추고 있다. 
* 부드러운 라인을 가지고 있다.

```
df['분양가'].plot(kind='kde')
```

## Hexbin

* hexbin은 고밀도 산점도 그래프.
* x와 y 키 값을 넣어 주어야 한다.
* x, y 값 모두 numeric 한 값
* 데이터의 밀도를 추정

```
df.plot(kind='hexbin', x='분양가', y='연도', gridsize=20)
```

## 박스 플롯(box)

```
df_seoul = df.loc[df['지역'] == '서울']
```

```
df_seoul['분양가'].plot(kind='box')
```

```
from IPython.display import Image

# image source : https://justinsighting.com/how-to-interpret-box-plots/
Image('https://justinsighting.com/wp-content/uploads/2016/12/boxplot-description.png')
```

```
df_seoul.describe()
```

IQR은 Inter Quantile Range의 약어로써, (3Q - 1Q) * 1.5 값

```
IQR = (7732 - 6519.75) * 1.5
IQR
```

```
## 박스 플롯 max
7732 + IQR
```

```
## 박스 플롯 min
6519.75 - IQR
```

박스플롯은 데이터 outlier 감지 할 떄 가장 많이 활용 되며, 25%, median, 75% 분위의 값을 활용하는 용도로 많이 활용된다.  
outlier가 오류의 값이거나, 등등의 이유로 배제하여 데이터 분석하기 위함.
또는 뭔가 다른 시각으로 접근해야하는지 점검하기 위함.

## area plot

* area plot은 line 그래프에서 아래 area를 모두 색칠해 주는 것이 특징  


```
df.groupby('월')['분양가'].count().plot(kind='line')
```

```
df.groupby('월')['분양가'].count().plot(kind='area')
```

## pie plot (파이 그래프)

pie는 대표적으로 데이터의 점유율을 보여줄 때 

```
df.groupby('연도')['분양가'].count().plot(kind='pie')
```

## scatter plot (산점도 그래프)

* 점으로 데이터를 표기
* x, y 값을 넣어주어야 한다 (hexbin과 유사)
* x축과 y축을 지정해주면 그에 맞는 데이터 분포도를 볼 수 있다.
* 역시 numeric 한 column 만 지정할 수 있다

```
df.plot(x='월', y='분양가',kind='scatter')
```
