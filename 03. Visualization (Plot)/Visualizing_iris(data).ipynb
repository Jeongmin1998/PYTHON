# Visualizing


```python
import sklearn
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=load_iris() #iris 데이터 불러오기
iris.keys()
```

```python
iris
```

```python
df=pd.DataFrame(iris['data'],columns=iris['feature_names']) #데이터프레임으로 만들어주기
df.head()
```

```python
df['label']=iris['target'] #컬럼 추가
df.head()
```

```python
df['name']= df['label'].apply(lambda x: iris['target_names'][x])
df.head()
```

## Swarmplot



**Grid: 플롯을 읽고 특정 값에 대한 정보를 명확하게 알 수 있도록.  
  
### seaborn에서 set_style이라는 명령으로 전체적인 스타일을 조절

```python
sns.set(style="whitegrid")
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.swarmplot(x="name", y="petal length (cm)",data=df)
```

#### 그래프를 가로로 눕히고 싶다면? x, y 바꿔주기!

```python
sns.set(style="whitegrid")
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.swarmplot(x="petal length (cm)", y="name",data=df)
```

```python
sns.set(style="whitegrid")
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.swarmplot(x="name", y="petal width (cm)",data=df)
```

```python
sns.set(style="whitegrid")
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.swarmplot(x="name", y="sepal length (cm)",data=df)
```

```python
sns.set(style="whitegrid")
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.swarmplot(x="name", y="sepal width (cm)",data=df) #구분 불명확
```

## Violinplot


**- Despine(축/테두리 제거)
디폴트값은 네 개의 spine이 모두 있음; left, right, top, bottom**

(기본적으로 top, right 두 개를 제거)

```python
sns.set(style='whitegrid')
g= sns.violinplot(data=df, x="name", y="petal length (cm)")
sns.despine(left=True)
```

## Heatmap

**annot=True 각 셀에 숫자 써주라는 arguement**

```python

sns.heatmap(df.corr(), annot=True, cmap='Pastel1')
```

## Pairplot

```python
sns.pairplot(data=df, kind='scatter')
```

## Boxplot


```python
sns.set(style="ticks", palette="Pastel1") #ticks 눈금자 없애는 스타일
sns.boxplot(x="name", y="petal length (cm)", palette=["m", "g"],
            data=df)
sns.despine(offset=10, trim=True)
```

## Barplot

```python
sns.set(style='whitegrid')
sns.catplot(data=df, kind="bar", x="name", y="petal length (cm)", palette="dark", alpha=.6, height=6)
sns.despine(left=True)
plt.xlabel("NAME")
plt.ylabel("Petal Length (cm)")
plt.title("PL Barplot")
```

## Lineplot

```python
sns.set(style="darkgrid")
sns.lineplot(data=df, x="name", y="petal length (cm)")
```
